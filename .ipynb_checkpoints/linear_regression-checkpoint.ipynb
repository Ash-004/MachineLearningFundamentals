```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def generateDataform(m):
    X = np.random.randn(m)*10
    noise = np.random.randn(m)
    y = 3*X+1 + noise * 5
    return X,y
    
X,y = generateDataform(100)
```

```python
def plotData(X,y,color='orange'):
    plt.scatter(X,y,color=color)
    plt.show()
```

```python
plotData(X,y,'red')
```

```python
def normalizeData(X):
    X = (X-X.mean())/X.std()
    return X
```

```python
X = normalizeData(X)
plotData(X,y)
```

```python
def trainTestSplit(X,y,split=0.8):
    m = X.shape[0]
    data = np.zeros((m,2))
    data[:,0] = X
    data[:,1] = y
    
    np.random.shuffle(data)
    split = int(m*split)
    
    XT = data[:split,0]
    yT = data[:split,1]
    
    Xt = data[split:,0]
    yt = data[split:,1]
    
    return XT,yT,Xt,yt
```

```python
XT,yT,Xt,yt = trainTestSplit(X,y)
```

```python
plt.scatter(XT,yT,color="black",label="Train Data")
plt.scatter(Xt,yt,color="yellow",label="Test Data")
plt.legend()
plt.show()
```

```python
def hypothesis(X,theta):
    return theta[0]+theta[1]*X

def error(X,y,theta):
    m = X.shape[0]
    e = 0
    
    for i in range(m):
        y_i = hypothesis(X[i],theta)
        e = e + (y[i] - y_i)**2
        
    return e/(2*m)
```

```python
def gradient(X,y,theta):
    m = X.shape[0]
    grad = np.zeros((2,))
    
    for i in range(m):
        exp = hypothesis(X[i],theta) - y[i]
        grad[0] += (exp)
        grad[1] += (exp)*X[i]
        
    return grad/m
        

def train(X,y,learning_rate = 0.1):
    theta = np.zeros((2,))
    m = 100
    
    error_list = []
    for i in range(m):
        grad = gradient(X,y,theta)
        error_list.append(error(X,y,theta))
        theta[0] = theta[0] - learning_rate * grad[0]
        theta[1] = theta[1] - learning_rate * grad[1]
        
    plt.plot(error_list)
    return theta
```

```python
theta = train(X,y)
```

```python
theta
```

```python
def predict(X,theta):
    return hypothesis(X,theta)
     
yp = predict(Xt,theta)
```

```python
plt.scatter(XT,yT,label="train")
plt.scatter(Xt,yt,color='orange',label="test")
plt.plot(Xt,yp,color='green',label="prediction")
plt.legend()
plt.show()
```

```python
def r2score(y,yp):
    y_mean = y.mean()
    t1 = np.sum((y - yp)**2)
    t2 = np.sum((y - y_mean)**2)
    
    return (1 - (t1/t2))
r2score(yt,yp)
```
