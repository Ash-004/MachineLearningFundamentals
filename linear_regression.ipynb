```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def generateDataform(m):
    X = np.random.randn(m)*10
    noise = np.random.randn(m)*5
    y = 3*X+1 + noise
    return (X,y)
    
X,y = generateDataform(100)
```

```python
def plotData(X,y,color='orange'):
    plt.scatter(X,y,color=color)
    plt.show()
```

```python
plotData(X,y,'red')
```

```python
def normalizeData(X):
    X = (X-X.mean())/X.std()
    return X
```

```python
nX = normalizeData(X)
plotData(nX,y)
```

```python
def trainTestSplit(X,y,split=0.8):
    m = X.shape[0]
    data = np.zeros((m,2))
    data[:,0] = X
    data[:,1] = y
    
    np.random.shuffle(data)
    split = int(m*split)
    
    XT = data[:split,0]
    yT = data[:split,1]
    
    Xt = data[split:,0]
    yt = data[split:,1]
    
    return XT,yT,Xt,yt
```

```python
XT,yT,Xt,yt = trainTestSplit(X,y)
```

```python
plt.scatter(XT,yT,color="black",label="Train Data")
plt.scatter(Xt,yt,color="yellow",label="Test Data")
plt.legend()
plt.show()
```
