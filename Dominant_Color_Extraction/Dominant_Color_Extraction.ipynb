```python
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
import numpy as np
```

```python
img = cv2.imread("Images/Landscape-Color.jpg")
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
img = cv2.resize(img,(382,306))
```

```python
plt.imshow(img)
plt.show()
```

```python
X = img.reshape((-1,3))
print(X.shape)
```

```python
k = 5
model = KMeans(n_clusters=k)
model.fit(X)
```

```python
centroids = model.cluster_centers_
```

```python
colors = np.array(centroids,dtype = 'uint8')
```

```python
i = 1
for color in colors:
    plt.subplot(1,k,i)
    plt.axis("off")
    i+=1
    mat = np.zeros((100,100,3),dtype='uint8')
    mat[:,:,:] = color
    plt.imshow(mat)
    plt.show()
```
